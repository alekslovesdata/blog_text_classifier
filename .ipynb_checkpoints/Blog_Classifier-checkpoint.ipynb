3. Now let's make sure the groups are balanced

We want to end up with about 50,000 in science and technology and 50,000 in other

```python
#create new column for binary grouping
def binary_topic(row):
    if row['topic'] == 'Technology':
        group = 'tech_and_science'
    elif row['topic'] == 'Science':
        group = 'tech_and_science'
    else:
        group = 'other'
    return group

data_30s['binary_topic'] = data.apply(binary_topic, axis=1)
data_30s.head()
```

```python
#Typically we would use a SMOTE anlaysis or randomized oversampling, but to save time, demo with simple random undersampling
#If we have more time, oversample using Tf-Idf representation of texts

undersampled_other = data_30s[data_30s['binary_topic'] == 'other'].sample(n = 9750)
technology_and_science = data_30s[data_30s['binary_topic'] == 'tech_and_science']

#now paste them back together
balanced_data = pd.concat([undersampled_other, technology_and_science], ignore_index=True)
balanced_data['binary_topic'].value_counts()
        
```

```python
# Making a custom text cleaner to get rid of contractions, apostrophes, mispellings, pronouns, special characters, and punctuation
# Install necessary packages in terminal, including python3 -m spacy download en

import codecs
import unidecode
import re
import spacy
nlp = spacy.load("en")

def spacy_cleaner(text):
    try:
        decoded = unidecode.unidecode(codecs.decode(text, 'unicode_escape'))
    except:
        decoded = unidecode.unidecode(text)
    apostrophe_handled = re.sub("’", "'", decoded)
    expanded = ' '.join([contraction_mapping[t] if t in contraction_mapping else t for t in apostrophe_handled.split(" ")])
    parsed = nlp(expanded)
    final_tokens = []
    for t in parsed:
        if t.is_punct or t.is_space or t.like_num or t.like_url or str(t).startswith('@'):
            pass
        else:
            if t.lemma_ == '-PRON-':
                final_tokens.append(str(t))
            else:
                sc_removed = re.sub("[^a-zA-Z]", '', str(t.lemma_))
                if len(sc_removed) > 1:
                    final_tokens.append(sc_removed)
    joined = ' '.join(final_tokens)
    spell_corrected = re.sub(r'(.)\1+', r'\1\1', joined)
    return spell_corrected
```

5. Label Coding

```python
label_codes = {
    'tech_and_science': 0,
    'other': 1,
}
```

```python
balanced_data['label'] = [0 if x =='tech_and_science' else 1 for x in balanced_data['binary_topic']]
balanced_data.head()
```

6. Train Test Split

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import chi2
```

```python
X_train, X_test, y_train, y_test = train_test_split(balanced_data['clean_text'],
balanced_data['label'], test_size = 0.15, random_state = 8)
```

7. Text Representation  


I'll use TF-IDF Vectors as features

```python
# Parameter selection
ngram_range = (1,2)
min_df = 15
max_df = 1.
max_features = 400
```

```python
tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=ngram_range,
                        stop_words=None,
                        lowercase=False,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2',
                        sublinear_tf=True)
                        
features_train = tfidf.fit_transform(X_train).toarray()
labels_train = y_train
print(features_train.shape)

features_test = tfidf.transform(X_test).toarray()
labels_test = y_test
print(features_test.shape)
```

```python
#Now check which words are most correlated to each category

from sklearn.feature_selection import chi2
import numpy as np

for Product, category_id in sorted(label_codes.items()):
    features_chi2 = chi2(features_train, labels_train == category_id)
    indices = np.argsort(features_chi2[0])
    feature_names = np.array(tfidf.get_feature_names())[indices]
    unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
    bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
    print("# '{}' category:".format(Product))
    print("  . Most correlated unigrams:\n. {}".format('\n. '.join(unigrams[-5:])))
    print("  . Most correlated bigrams:\n. {}".format('\n. '.join(bigrams[-2:])))
    print("")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from pprint import pprint
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import ShuffleSplit
```

```python

rf_0 = RandomForestClassifier(random_state = 8)

print('Parameters currently in use:\n')
pprint(rf_0.get_params())
```

```python
#define the grid
# n_estimators
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 5)]

# max_features
max_features = ['auto', 'sqrt']

# max_depth
max_depth = [int(x) for x in np.linspace(20, 100, num = 5)]
max_depth.append(None)

# min_samples_split
min_samples_split = [2, 5, 10]

# min_samples_leaf
min_samples_leaf = [1, 2, 4]

# bootstrap
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

pprint(random_grid)
```

```python
# First create the base model to tune
rfc = RandomForestClassifier(random_state=8)

# Definition of the random search
random_search = RandomizedSearchCV(estimator=rfc,
                                   param_distributions=random_grid,
                                   n_iter=50,
                                   scoring='accuracy',
                                   cv=3, 
                                   verbose=1, 
                                   random_state=8, n_jobs = 12)

# Fit the random search model
random_search.fit(features_train, labels_train)
```

```python
print("The best hyperparameters from Random Search are:")
print(random_search.best_params_)
print("")
print("The mean accuracy of a model with these hyperparameters is:")
print(random_search.best_score_)
```

```python
# Given more time Create the parameter grid based on the results of random search 
```

```python
best_rfc = random_search.best_estimator_
best_rfc.fit(features_train, labels_train)
```

```python
rfc_pred = best_rfc.predict(features_test)
```

# 8. Training and Test Accuracy

```python
print("The training accuracy is: ")
print(accuracy_score(labels_train, best_rfc.predict(features_train)))
print("The test accuracy is: ")
print(accuracy_score(labels_test, rfc_pred))
```

```python
# Classification report
print("Classification report")
print(classification_report(labels_test,rfc_pred))
```

# 9. Refine the current model to improve the test accuracy

```python
#Maybe go back to the text cleaning part and remove common words and do a few more text cleaning tricks?
balanced_data['clean_text'] = balanced_data['clean_text'].str.replace("\r", " ")
balanced_data['clean_text'] = balanced_data['clean_text'].str.replace("\n", " ")
balanced_data['clean_text'] = balanced_data['clean_text'].str.replace("    ", " ")
balanced_data['clean_text'] = balanced_data['clean_text'].str.replace('"', '')

#make lowercase
balanced_data['clean_text'] = balanced_data['clean_text'].str.lower()

#remove punctuation
punctuation_signs = list("?:!.,;")
for punct_sign in punctuation_signs:
    balanced_data['clean_text'] = balanced_data['clean_text'].str.replace(punct_sign, '')
    
#remove possessive pronouns
balanced_data['clean_text'] = balanced_data['clean_text'].str.replace("'s", "")
```
